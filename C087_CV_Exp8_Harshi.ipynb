```python
SIZE = 25
SIGMA = 2
WAVELENGTH = 5
THETA = np.pi/2
PSI = 0
GAMMA = 0.5

gaborFilter = cv2.getGaborKernel((SIZE, SIZE), SIGMA, THETA, WAVELENGTH, GAMMA, PSI)
plt.imshow(gaborFilter, cmap="gray")
```

```python
gaborFilter= gaborFilter/gaborFilter.sum()
```

```python
image1_f=cv2.filter2D(img1_g,-1,gaborFilter)
plt.imshow(image1_f,cmap='gray')
```

```python
plt.imshow(img2)
```

```python
filters=[]
num_filters=16
SIZE = 31
SIGMA = 2
WAVELENGTH = 5
PSI = 0
GAMMA = 0.5
```

```python
for theta in np.arange(0, np.pi, np.pi / num_filters):
    f = cv2.getGaborKernel((SIZE, SIZE), SIGMA, theta, WAVELENGTH, GAMMA, PSI)
    f=f/f.sum()
    filters.append(f)
```

```python
image2_f=np.zeros_like(img2)
```

```python
for f in filters:
    image2_f=cv2.filter2D(img2,-1,f)
    np.maximum(image2_f,img2,image2_f)
```

```python
plt.subplot(1,2,1)
plt.imshow(img2)
plt.title('Orignal Image')
plt.subplot(1,2,2)
plt.imshow(image2_f)
plt.title('Filtered Image with Edges')
```

Conclusion: for the given image elephant, 16 filters in the range of 0 to pi are used. The filtered image shows the highlighted edges in 16 directions. If the number of filters is reduced to 4 then the filtered output then the filtered image shows vertical,backward, horizontal,upward direction.
